```python
import numpy as np
import pandas as pd
import random
```

```python
'''
Dados de teste que servirão para popular o banco de dados
'''
df_nomes     = pd.read_csv("Dados Banco de Dados Denuncia - Nomes e sobrenomes.csv")
df_enderecos = pd.read_csv("Dados Banco de Dados Denuncia - Endereços.csv")
df_ocorrencias = pd.read_csv("Dados Banco de Dados Denuncia - Ocorrências.csv")
```

## Objetivo - Gerar 1300 denúncias

- **38%** Violência Doméstica      (tipo 1) >> 494
- **32%** Violência Física         (tipo 2) >> 416
- **17%** Violência Psicologica    (tipo 3) >> 221
- **8%**  Violência Sexual         (tipo 4) >> 104
- **2%**  Cárcere Privado          (tipo 6) >>  26

Aleḿ disso, ao gerar cada denúncia terá uma chance de:
- 40% de ela estar aguardando avaliação
- 25% em avaliação
- 20% de ela já ter sido encaminhada
- 15% de ela já ter sido arquivada.

```python
def gerarDenuncia(cat, numAcompanhamento):
    denuncia = {}
    # DADOS DA VITIMA
    denuncia['den_nome_vitima'] = df_nomes['Feminino'][random.randint(0, df_nomes['Feminino'].count()-1)]
    denuncia['den_nome_vitima'] += " " + df_nomes['Sobrenome 1'][random.randint(0, df_nomes['Sobrenome 1'].count()-1)]
    denuncia['den_nome_vitima'] += " " + df_nomes['Sobrenome 2'][random.randint(0, df_nomes['Sobrenome 2'].count()-1)]
    
    idEndVitima   = random.randint(0, df_enderecos['Endereço'].count()-1)
    denuncia['den_endereco_vitima'] = df_enderecos['Endereço'][idEndVitima]
    denuncia['den_endereco_vitima'] += " nº " + str(random.randint(0, 2000)) + ","
    denuncia['den_endereco_vitima'] += " " + df_enderecos['Bairro'][idEndVitima]
    denuncia['den_endereco_vitima'] += " " + df_enderecos['Cidade'][idEndVitima]
    
    denuncia['fae_id_vitima']       = str(random.randint(2, 4));
    
    # DADOS DO SUSPEITO
    denuncia['den_nome_suspeito'] = df_nomes['Masculino'][random.randint(0, df_nomes['Masculino'].count()-1)]
    denuncia['den_nome_suspeito'] += " " + df_nomes['Sobrenome 1'][random.randint(0, df_nomes['Sobrenome 1'].count()-1)]
    denuncia['den_nome_suspeito'] += " " + df_nomes['Sobrenome 2'][random.randint(0, df_nomes['Sobrenome 2'].count()-1)]
    
    idEndSuspeito = random.randint(0, df_enderecos['Endereço'].count()-1)
    denuncia['den_endereco_suspeito'] = df_enderecos['Endereço'][idEndSuspeito]
    denuncia['den_endereco_suspeito'] += " nº " + str(random.randint(0, 2000)) + ","
    denuncia['den_endereco_suspeito'] += " " + df_enderecos['Bairro'][idEndSuspeito]
    denuncia['den_endereco_suspeito'] += " " + df_enderecos['Cidade'][idEndSuspeito]
    
    denuncia['fae_id_suspeito']       = str(random.randint(2, 4));
    
    # DADOS DA OCORRENCIA
    idDenuncia = random.randint(0, df_ocorrencias[df_ocorrencias['Tipo de denúncia'] == cat]['Tipo de denúncia'].count()-1)
    
    denuncia['con_id']               = df_ocorrencias['Conexao'][idDenuncia]
    denuncia['den_local_ocorrencia'] = df_ocorrencias['Local da agressão'][idDenuncia]
    denuncia['den_desc_ocorrencia']  = df_ocorrencias['Mensagem'][idDenuncia]
    denuncia['den_uf_ocorrencia']    = "SP"
    denuncia['cat_id']               = cat
    
    denuncia['den_cd_acompanhamento'] = ("XXXXXXXXXXX" + str(numAcompanhamento%100000).zfill(5))
    # denuncia['den_cd_acompanhamento'] = idDenuncia
    
    # STATUS DA OCORRENCIA
    randStatus = random.randint(0, 100)
    if randStatus <= 40:
        denuncia['sta_id'] = '1'
    elif randStatus <= 65:
        denuncia['sta_id'] = '2'
    elif randStatus <= 85:
        denuncia['sta_id'] = '3'
    else:
        denuncia['sta_id'] = '4'
    
    denuncia['den_dt_criacao']     = "NOW()"
    denuncia['den_dt_modificacao'] = "NOW()"
    
    return denuncia
```

## Gerando um único dado para testar inserção csv

```python
df_denuncias = pd.DataFrame(columns=['den_nome_vitima', 'den_endereco_vitima', 'fae_id_vitima', 
                                      'den_nome_suspeito', 'den_endereco_suspeito', 'fae_id_suspeito','con_id',
                                      'den_local_ocorrencia', 'den_desc_ocorrencia', 'den_uf_ocorrencia', 'cat_id',
                                      'den_cd_acompanhamento', 'den_dt_criacao', 'den_dt_modificacao'])

contador = 1

# Para adicionar uma denuncia
# df_denuncias = df_denuncias.append(gerarDenuncia(1, 10), ignore_index=True)
```

```python
df_denuncias.shape
```

```python
# 38% Violência Doméstica (tipo 1) >> 494
for i in range(0, 495):
    df_denuncias = df_denuncias.append(gerarDenuncia(1, contador), ignore_index=True)
    contador += 1
    
print(contador)
```

```python
# 32% Violência Física (tipo 2) >> 416
for i in range(0, 417):
    df_denuncias = df_denuncias.append(gerarDenuncia(2, contador), ignore_index=True)
    contador += 1

print(contador)
```

```python
# 17% Violência Psicologica (tipo 3) >> 221
for i in range(0, 222):
    df_denuncias = df_denuncias.append(gerarDenuncia(3, contador), ignore_index=True)
    contador += 1

print(contador)
```

```python
# 8% Violência Sexual (tipo 4) >> 104
for i in range(0, 105):
    df_denuncias = df_denuncias.append(gerarDenuncia(4, contador), ignore_index=True)
    contador += 1
    
print(contador)
```

```python
# 2% Cárcere Privado (tipo 6) >> 26
for i in range(0, 27):
    df_denuncias = df_denuncias.append( gerarDenuncia(6, contador), ignore_index=True)
    contador += 1
print(contador)
```
